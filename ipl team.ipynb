```python
totalrun= df1.groupby('batter').agg({'batsman_run':'sum','batter':'count'})
totalrun.rename(columns={'batter': 'No of bolles'},inplace=True)
totalrun
```

```python
top5batsman= totalrun.sort_values(by='batsman_run',ascending=False).head(5)
top5batsman
```

```python
totalrun['stickrate'] = totalrun['batsman_run']/totalrun['No of bolles']*100
totalrun
```

```python
#the strikerate of all players
totalrun.loc[totalrun['No of bolles'] >= 60, 'best stickrate'] =totalrun['stickrate']
top5stickrate= totalrun.sort_values(by='best stickrate',ascending=False).head(5)
top5stickrate
```

```python
#these are the top 5 strikers who played atleast 60 balls
total=df1.groupby('batter').agg({'batsman_run':'sum','isWicketDelivery':'sum'})
total.rename(columns={'isWicketDelivery': 'No of times out'},inplace=True)
total
```

```python
total.loc[total['No of times out'] == 0, 'batting average'] = total['batsman_run']
total.loc[total['No of times out'] != 0, 'batting average'] = (total['batsman_run']/total['No of times out'])
total
```

```python
total.loc[total['batsman_run'] >= 200, 'best batting average'] =total['batting average']
top5= total.sort_values(by='best batting average',ascending=False).head(5)
top5
```

```python
#top 5 average batsman who scored at least 200 runs
totalball= df1.groupby('bowler').agg({'total_run':'sum','bowler':'count','isWicketDelivery':'sum'})
totalball.rename(columns={'total_run': 'No of run','bowler':'no of ball','isWicketDelivery':'wickets'},inplace=True)
totalball
```

```python
totalball['bowling economy rate'] = totalball['No of run']/totalball['no of ball']*6
totalball
```

```python
totalball.loc[totalball['no of ball'] >= 60, 'best bowling economy rate'] =totalball['bowling economy rate']
top5b= totalball.sort_values(by='best bowling economy rate').head(5)
top5b
```

```python
#top5 wicket taking bowlers
totalball['wickets economy rate'] = totalball['wickets']/totalball['no of ball']
totalball.loc[totalball['no of ball'] >= 60, 'best wickets economy rate'] =totalball['wickets economy rate']
bestwicketer=totalball.sort_values(by='best wickets economy rate',ascending=False).head(5)
bestwicketer
```
